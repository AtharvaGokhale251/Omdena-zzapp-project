```
#Original Formula from paper:

#Risk parameter = [(Temp × 0.224) + (precipitation × 0.208) + (altitude × 0.104) + (slope × 0.073) + LULC × 0.082) + (DTWB × 0.123) + (DTR × 0.038) + 
#                   (Pop dens × 0.051) + (Mal prev × 0.051) + (NDVI × 0.047)]


#Formula used according to data available:

#Risk parameter = [(Temp × 0.224) + (precipitation × 0.208)]
#Due to changed formula, the threshold values for categorization will change accordingly


def riskp_calc(row):
  if (row['location']=='Amhara'):
    riskp = ((row['temp'] * 0.224) + (sumpr_amhara[str (row['year'])] * 0.208))
  elif (row['location']=='Obuasi'):
    riskp = ((row['temp'] * 0.224) + (sumpr_obuasi[str (row['year'])] * 0.208))
  elif (row['location']=='Eldoret'):
    riskp = ((row['temp'] * 0.224) + (sumpr_eldoret[str (row['year'])] * 0.208))
  elif (row['location']=='Kisumu'):
    riskp = ((row['temp'] * 0.224) + (sumpr_kisumu[str (row['year'])] * 0.208))
  elif (row['location']=='Malindi'):
    riskp = ((row['temp'] * 0.224) + (sumpr_malindi[str (row['year'])] * 0.208))
  elif (row['location']=='Mombasa'):
    riskp = ((row['temp'] * 0.224) + (sumpr_mombasa[str (row['year'])] * 0.208))
  elif (row['location']=='Nairobi'):
    riskp = ((row['temp'] * 0.224) + (sumpr_nairobi[str (row['year'])] * 0.208))
  elif (row['location']=='Nakuru'):
    riskp = ((row['temp'] * 0.224) + (sumpr_nakuru[str (row['year'])] * 0.208))
  elif (row['location']=='Thika'):
    riskp = ((row['temp'] * 0.224) + (sumpr_thika[str (row['year'])] * 0.208))

  return riskp
```

```
all_data['risk_p'] = all_data.apply(riskp_calc, axis=1)
```

```
def find_risk_status(row):
  if (row['risk_p'] < 98.528):
    risk_status = "low"
  elif ((row['risk_p'] >= 98.528) and (row['risk_p'] <= 151.872)):
    risk_status = "high"
  else:
    risk_status = "medium"

  return risk_status
```

```
all_data['risk_status'] = all_data.apply(find_risk_status, axis=1)
```

```
#Import libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn import decomposition
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter
```

```
Y = all_data2['risk_status'] #Choose Target response
X = all_data2.drop(['risk_status', 'date'],axis=1)
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=2)
```

```
pca = decomposition.PCA(n_components=4)
pca.fit(X_train)
X_train = pca.transform(X_train)
```

```
estModel = RandomForestRegressor(n_estimators = 30, random_state = 0)
estModel.fit(X_train, Y_train)
#Model Score
mScore = estModel.score(X_train,Y_train)
#Prediction
Predict_Y = estModel.predict(X_test)
```

```
rmse = mean_squared_error(Y_test,Predict_Y)
print('Root Mean Squared Error: %.2f' % np.sqrt(rmse))
#Variance score: 1 is perfect prediction score
print('Test Variance Score: %.4f' % r2_score(Y_test, Predict_Y))
#Variance score: 1 is perfect prediction score
print('R^2 Prediction Score: %.4f'% mScore)
```

```
fig, pX = plt.subplots()

pX.scatter(Y_test, np.round(Predict_Y, decimals=1), edgecolors=(0, 0, 0))
pX.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--', lw=2)
pX.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
pX.xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
pX.set_xlabel('Actual Data')
pX.set_ylabel('Predicted Data')
pX.set_title("Verified vs Predicted")
plt.show()
```
